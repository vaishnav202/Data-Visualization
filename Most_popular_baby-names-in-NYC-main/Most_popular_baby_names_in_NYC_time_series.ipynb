```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
baby_names_df = pd.read_csv("Most_Popular_Baby_Names_by_Sex_and_Mother_s_Ethnic_Group__New_York_City.csv")

# Display the first few rows of the dataset
print(baby_names_df.head())

# Aggregate the data by BRTH_YR
yearly_counts = baby_names_df.groupby('BRTH_YR')['CNT'].sum().reset_index()

# Plot the total count of baby names per year
plt.figure(figsize=(10, 6))
plt.plot(yearly_counts['BRTH_YR'], yearly_counts['CNT'], marker='o')
plt.title('Total Count of Baby Names per Year')
plt.xlabel('Year')
plt.ylabel('Total Count of Baby Names')
plt.grid(True)
plt.show()



```

```python
# Aggregate the data by BRTH_YR to get the number of unique names per year
unique_names_per_year = baby_names_df.groupby('BRTH_YR')['NM'].nunique().reset_index()

# Plot the number of unique names per year
plt.figure(figsize=(10, 6))
plt.plot(unique_names_per_year['BRTH_YR'], unique_names_per_year['NM'], marker='o', color='g')
plt.title('Number of Unique Names per Year')
plt.xlabel('Year')
plt.ylabel('Number of Unique Names')
plt.grid(True)
plt.show()
```

```python
# Define a function to plot trends for specific names
def plot_name_trend(name):
    name_trend = baby_names_df[baby_names_df['NM'] == name].groupby('BRTH_YR')['CNT'].sum().reset_index()
    plt.figure(figsize=(10, 6))
    plt.plot(name_trend['BRTH_YR'], name_trend['CNT'], marker='o')
    plt.title(f'Trend for Name: {name}')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.grid(True)
    plt.show()

# Example: Plot trend for the name "Emma"
plot_name_trend('Emma')
```

```python
# Aggregate the data by BRTH_YR and GNDR to get the total number of births per gender per year
gender_trends = baby_names_df.groupby(['BRTH_YR', 'GNDR'])['CNT'].sum().unstack()

# Plot the trends based on gender
gender_trends.plot(figsize=(10, 6), marker='o')
plt.title('Total Number of Births per Gender per Year')
plt.xlabel('Year')
plt.ylabel('Total Number of Births')
plt.grid(True)
plt.show()
```

```python
# Aggregate the data by BRTH_YR and ETHCTY to get the total number of births per ethnic group per year
ethnic_trends = baby_names_df.groupby(['BRTH_YR', 'ETHCTY'])['CNT'].sum().unstack()

# Plot the trends based on ethnic groups
ethnic_trends.plot(figsize=(10, 6), marker='o')
plt.title('Total Number of Births per Ethnic Group per Year')
plt.xlabel('Year')
plt.ylabel('Total Number of Births')
plt.legend(title='Ethnic Group')
plt.grid(True)
plt.show()
```
